```python
import json
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# import spline
from scipy.interpolate import splrep, BSpline

qm9_results = (
    "../../../../data/learning_experiments_1219/qm9/qm9_learning_results_dict.json"
)
qm9_results = json.load(open(qm9_results, "r"))
qm8_results = (
    "../../../../data/learning_experiments_1219/qm8/qm8_learning_results_dict.json"
)
qm8_results = json.load(open(qm8_results, "r"))
libe_results = (
    "../../../../data/learning_experiments_1219/libe/libe_learning_results_dict.json"
)
libe_results = json.load(open(libe_results, "r"))
```

```python
partitions = ["non_qtaim", "non_qtaim_bl", "qtaim", "qtaim_bl"]
data_sizes = [10, 100, 1000, 10000, 100000]


# iterate through partitions and sizes then plot
for part in partitions:
    temp_r2 = []
    for size in data_sizes:
        temp_r2.append(qm9_results[part + "_" + str(size)]["r2_manual"])
    # plt.plot(data_sizes, temp_r2, label=part)
    # plot with some smoothing
    x = np.array(data_sizes)
    y = np.array(temp_r2)
    # spl = splrep(x, y, k=3)
    # y_smooth = BSpline(x, y, x_new)
    # plt.plot(x_new, y_smooth, label=part)
    plt.plot(x, y, label=part)
# log scale x
plt.xscale("log")
plt.title("QM9 - Learning Curve")
plt.xlabel("Training Data Size")
plt.ylabel("Test R2 (4-task average)")
plt.legend()
# increase dpi
plt.savefig("qm9_learning_curve.png", dpi=300)

plt.show()
plt.close()
```
